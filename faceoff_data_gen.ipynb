```python
%matplotlib inline
import matplotlib.pyplot as plt

from prepare_data import prepare_data, augment_data
face_data = prepare_data('/Users/azhong/face/clmtools/pdm_builder/data/images/',
                         '/Users/azhong/face/clmtools/pdm_builder/data/annotations.csv',
                         'muct_clmtools')
```

```python
a = face_data[0].image.copy()
x1, y1, x2, y2 = face_data[0].get_bounding_box('face')
import cv2
cv2.circle(a, (x1, y1), 10, (255), -1)
cv2.circle(a, (x2, y2), 10, (255), -1)
plt.imshow(a, cmap='gray')
```

```python
b = augment_data(face_data, 10, 'mouth', 10, 'face', 128, 0.1, 'mouth', 0.2, 0.2, 'mouth', (64, 64))
# b = augment_data(face_data, 10, 'left_eye', 10, 'face', 128, 0.1, 'left_eye', 0.2, 0.2, 'left_eye', (32, 32))
# b = augment_data(face_data, 10, 'right_eye', 10, 'face', 128, 0.1, 'right_eye', 0.2, 0.2, 'right_eye', (32, 32))
```

```python
for bb in b:
    bbi = bb.image.copy()
    for point in bb.landmarks:
        cv2.circle(bbi, (int(point[0]), int(point[1])), 1, (255), -1)
    plt.imshow(bbi, cmap='gray')
    plt.show()
```
