```python
csv_in = '/Users/azhong/face/clmtools/pdm_builder/data/muct-landmarks/muct76-opencv.csv'
image_path = '/Users/azhong/face/clmtools/pdm_builder/data/images/'
with open(csv_in) as fi:
    header = True
    line_count = 0
    line_limit = 1
    landmark_count = 76
    for line in fi:
        if header:
            header = False
            print(line)
            continue
        if line_count >= line_limit:
            break
            
        splitted = line.split(',')
        filename = splitted[0]
        full_path = image_path + filename + '.jpg'
        image = cv2.cvtColor(cv2.imread(full_path), cv2.COLOR_BGR2GRAY)
        
        landmarks = [(float(splitted[i*2+2]), float(splitted[i*2+3])) for i in range(landmark_count)]
        for i in range(landmark_count):
            cv2.circle(image, (int(landmarks[i][0]), int(landmarks[i][1])), 2, 255, -1)
        plt.imshow(image, cmap='gray')
        plt.show()
        line_count += 1
        print(line)
```
